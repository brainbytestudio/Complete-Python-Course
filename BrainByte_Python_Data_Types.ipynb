## üìå Why Data Types Matter
Every value in Python has a data type. Knowing how data types behave is **fundamental** to writing effective Python code.
In Python, the data type of a variable is determined at runtime, so there's no need to declare it explicitly beforehand‚Äîthis is why Python is known as a dynamically-typed language. A data type defines the kind of value a variable holds and what operations can be performed on it. Since everything in Python is an object, data types are essentially classes, and variables are instances (objects) of those classes.



## üî¢ 1. Numeric Data Types
Python supports three main numeric types:
- `int`: Integer values
- `float`: Decimal values
- `complex`: Complex numbers

### ‚ûï Integers (`int`)
Integers are whole numbers, positive or negative, without decimals.

```python
a = 10
b = -300
print(type(a), type(b))
```

### ‚ûó Floats (`float`)
Floats are numbers with decimal points.

```python
pi = 3.14159
gravity = 9.8
print(type(pi), type(gravity))
```

### üîÄ Complex Numbers (`complex`)
Complex numbers are written with a `j` as the imaginary part.

```python
z = 3 + 4j
print(type(z), z.real, z.imag)
```

## üî§ 2. Text Type - `str`
Strings are **immutable** ‚Äî once created, you cannot change individual characters. Any modification results in a new string object..

```python
name = "BrainByte"
quote = 'Learn Python, change your future.'
multi_line = """This is
a multi-line
string."""
print(name, type(name))
print(multi_line)
```

Useful string methods:

```python
print(name.upper())
print(quote.lower())
print("Byte" in name)
```

## üßÆ 3. Sequence Types

### üìã Lists (`list`)
**Mutable** means that the contents of the list can be changed after creation. You can add, remove, or modify elements.

```python
fruits = ["apple", "banana", "mango"]
fruits.append("grape")
print(fruits)
```

### üßä Tuples (`tuple`)
Tuples are **immutable**, ordered collections.

**Immutable** means that once the tuple is created, its contents cannot be changed. You cannot add, remove, or modify elements.

```python
dimensions = (1920, 1080)
print(dimensions)
```

### üìú Ranges (`range`)
Often used in loops.

```python
numbers = list(range(5))
print(numbers)
```

## üß† 4. Set Types
Set is an **unordered** collection of data types that is iterable, **mutable** and has no duplicate elements. The order of elements in a set is undefined though it may consist of various elements.
Sets are **mutable** ‚Äî you can add or remove elements, but since they are unordered, you cannot access them by index.

```python
unique_numbers = {1, 2, 3, 2, 1}
print(unique_numbers)  # duplicates removed
```

### Set operations:

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

## üóÉÔ∏è 5. Dictionary Type - `dict`
A dictionary in Python is an **unordered** collection of data values in form of **key-value pairs**. Key-value is provided in the dictionary to make it more optimized. Each key-value pair in a Dictionary is separated by a colon : , whereas each key is separated by a ‚Äòcomma‚Äô.

Dictionaries are **mutable** ‚Äî you can add new key-value pairs, update existing ones, or delete them.

```python
student = {
    "name": "Alice",
    "age": 22,
    "course": "Python"
}
print(student["name"])
print(student.get("age"))
```

You can add or update:

```python
student["grade"] = "A"
print(student)
```

## ‚úÖ 6. Boolean Type - `bool`
Booleans represent `True` or `False`.

```python
is_logged_in = True
print(type(is_logged_in))
print(5 > 3)
print(10 == 11)
```

## üßπ 7. None Type - `NoneType`
Represents the absence of a value.

```python
nothing = None
print(type(nothing))
```

## üß™ Type Checking and Type Casting
You can check and convert types:

```python
x = 5
print(type(x))
x = str(x)
print(type(x))
```
